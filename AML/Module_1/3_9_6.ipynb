```python
df.groupby(['driver_gender', 'driver_race'])[['is_arrested']].value_counts().unstack()
```

```python
df.groupby(['driver_gender', 'driver_race'])[['is_arrested']].value_counts(normalize=True).unstack()
```

На полученном срезе видно, что среди всех нарушителей значительно чаще (в 2-2.5 раза чаще) были арестованы нарушители рас `Hispanic` и `Black`. Причём данное утверждение справедливо как для мужчин, так и для женщин.
